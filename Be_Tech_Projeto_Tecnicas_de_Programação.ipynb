### 3 - Os 10 pa√≠ses com maior n√∫mero de casos por milh√£o de habitantes e os 10 pa√≠ses com maior n√∫mero de mortes por milh√£o de habitantes. Um dado est√° relacionado a outro?

```python
#Definir os 10 pa√≠ses com maior n√∫mero de casos por milh√£o de habitantes.
```

```python
#Manter apenas a √∫ltima linha de cada pa√≠s
covid_deaths_filter = covid_deaths_df.drop_duplicates(subset="location",
                                                      keep="last")
```

```python
#Colocar em ordem decrescente por n√∫mero de casos
cases_descending = covid_deaths_filter.sort_values(by="total_cases_per_million",
                                ascending=False)
```

```python
#Mostrar os 10 pa√≠ses com maior n√∫mero de casos por milh√£o de habitantes.
print(cases_descending.head(10)[
    ["location",
     "total_cases_per_million"]
    ])
```

```python
#Definir os 10 pa√≠ses com maior n√∫mero de mortes por milh√£o de habitantes.

```

```python
deaths_descending = covid_deaths_filter.sort_values(by="total_deaths_per_million",
                                ascending=False)
```

```python
#Mostrar os 10 pa√≠ses com maior n√∫mero de mortes por milh√£o de habitantes.
print(cases_descending.head(10)[
    ["location",
     "total_deaths_per_million"]
    ])
```

```python
#Nomeando as duas tabelas:

top_10_cases_milion = cases_descending.head(10)[
    ["location",
     "total_cases_per_million"]
    ]

top_10_deaths_milion = cases_descending.head(10)[
    ["location",
     "total_deaths_per_million"]
    ]
```

```python
#Juntando as duas tabelas geradas
df_merged = pd.merge(top_10_cases_milion, top_10_deaths_milion, how='inner', on='location')
```

```python
print(df_merged)
```

√â poss√≠vel perceber que TODOS os 10 pa√≠ses com maior n√∫mero de casos a cada um milh√£o de habitantes tamb√©m foram os pa√≠ses com maior n√∫mero de mortes a cada um milh√£o de habitantes. Correla√ß√£o de 100%.

### 4 - Qual continente foi mais afetado pelo Covid?

```python
# Filtrar os dados pela √∫ltima data de cada pa√≠s
last_date = covid_deaths_df.drop_duplicates(
    subset="location",
    keep="last")
```

```python
# Agrupar os dados por continente
continent = last_date.groupby(
    "continent")
```

```python
# Somar os valores da coluna total_deaths para cada grupo
deaths_continent = continent["total_deaths"].sum()
```

```python
# Resultado do n√∫mero de mortes por continente, em ordem crescente.

print(deaths_continent.sort_values(ascending=False))
```

O continente com maior n√∫mero de mortes foi a √Åsia, seguida pela Am√©rica do Sul, Europa, Am√©rica do Norte, √Åfrica e Oceania.

### 5 - Mostre o gr√°fico para a tabela criada na an√°lise anterior.

```python
#Tabela com os valores ordenados, para o gr√°fico ficar mais visual.
deaths_continent_sorted = deaths_continent.sort_values(ascending=False)
```

```python
#Criar o gr√°fico de barras com os eixos n√∫mero de mortes (Y) e continente (X).
plt.bar(deaths_continent_sorted.index, deaths_continent_sorted.values)
plt.title("Total de mortes por Covid por continente")
plt.xlabel("Continente")
plt.ylabel("Mortes")
plt.show()
```
